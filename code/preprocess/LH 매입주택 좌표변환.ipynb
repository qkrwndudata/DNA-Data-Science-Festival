# 주소변환

```python
#latitute
import requests, json
def get_lat(address):
  url = 'https://dapi.kakao.com/v2/local/search/address.json?query=' + address
  headers = {"Authorization": "KakaoAK b56426a9cd6d7bab39f8d654eb02a661"}
  api_json = json.loads(str(requests.get(url,headers=headers).text))
  address = api_json['documents'][0]['address']
  crd = dict(lat = address['y'], lng = address['x'])


  return address['y']
```

```python
#longitude
import requests, json
def get_lng(address):
  url = 'https://dapi.kakao.com/v2/local/search/address.json?query=' + address
  headers = {"Authorization": "KakaoAK b56426a9cd6d7bab39f8d654eb02a661"}
  api_json = json.loads(str(requests.get(url,headers=headers).text))
  address = api_json['documents'][0]['address']
  crd = dict(lat = address['y'], lng = address['x'])


  return address['x']
```

## 1. latitude

```python
#"서울특별시 강남구 봉은사로78길" 은 추출이 안 됨
lat1 = df.iloc[0:42,2].apply(get_lat)
lat1 = lat1.to_frame()
lat1.columns = ['lat']
print(lat1.shape)
lat1.head()
```

```python
lat2 = df.iloc[43:48,2].apply(get_lat)
lat2 = lat2.to_frame()
lat2.columns = ['lat']
print(lat2.shape)
lat2.head()
```

```python
lat = lat1.append(lat2)
print(lat.shape)
lat.head()
```

## longitude

```python
lng1 = df.iloc[0:42,2].apply(get_lng)
lng1 = lng1.to_frame()
lng1.columns = ['lng']

lng2 = df.iloc[43:48,2].apply(get_lng)
lng2 = lng2.to_frame()
lng2.columns = ['lng']

lng = lng1.append(lng2)
print(lng.shape)
lng.head()
```

## final dataframe

```python
df_final = pd.merge(df, lat, left_index = True, right_index = True, how='left')
df_final = pd.merge(df_final, lng, left_index = True, right_index = True, how='left')
df_final.head()
```

```python
df_final.to_csv('D:/graduate/DNA/data/LH 매입 행정동.csv')
```
