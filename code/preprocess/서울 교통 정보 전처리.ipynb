```python
from tqdm import tqdm
from collections import Counter

#주소명 통해 행정동 필터링
dong = [] #동 리스트
for i in tqdm(range(len(subway_df))):
    ###### 지하철역의 구, 도로명, 건물 번호 추출 ######
    address_letter = subway_df['ROAD_NM_ADDR'][i].split(' ')
    gu = address_letter[1] #구 설정
    if address_letter[3][-1] == '길': #길이 도로명에 붙어야 하는데 띄어쓰기가 되어있는 데이터가 있음. 이를 해결
        address_letter[2] = address_letter[2] + address_letter[3]
        del address_letter[3]
    road = address_letter[2] #도로명 설정
    if '-' in address_letter[3]: #건물번호 부번이 있을 경우
        building_no = address_letter[3].split('-')
        building_no = building_no[0]
    else:
        building_no = address_letter[3]
    ###### 우편번호 데이터랑 매칭 및 동 리스트에 행정동 추가######
    place = post_df[(post_df['시군구'] == gu) & (post_df['도로명'] == road) & (post_df['건물번호본번'] == int(building_no))]
    if len(place) == 0: #장소가 없어- 일단 결측치 채워넣어. 지하철 데이터는 인촌로인데 우편번호는 고려대로야...
        dong.append(np.nan)
    
    else:
        place = place.iloc[0]
        if pd.isna(place['행정동명']) : #### 행정동명이 없을 경우에는 같은 우편번호의 행정동을 가져오자, 
                                           #행정동이 여러개일 때는 최빈값 가져오기
            postno = place['우편번호']
            postno_df = post_df[post_df['우편번호'] == postno]
            postno_df = list(postno_df['행정동명'])
            newlist = [x for x in postno_df if pd.isnull(x) == False and x != 'nan']
            cnt = Counter(newlist)
            mode = cnt.most_common(1)
            name = mode[0][0]  
            dong.append(name)
            
        else:  #### 행정동명이 있을 경우에는 그냥 행정동명을 삽입
            name = place['행정동명']
            dong.append(name)
```

```python
subway_df['행정동'] = dong
```

```python
#행정동 결측치 어떻게든 메꿔보기 (인접 건물번호본번 가져오기)
for i in tqdm(list(bus_df[bus_df['행정동'].isna()].index)):
    if pd.isnull(bus_df.loc[i, '_ROAD_AD']): #그런데 도로명주소가 없는 경우에는 어쩔 수 없이 결측치
        bus_df.loc[i, '_ROAD_AD'] = np.nan

    else: #행정동이 결측치면


        address_letter = bus_df['_ROAD_AD'][i].split(' ')

        gu = address_letter[1] #구 설정
        if len(address_letter) == 4: #서울특별시 종로구 진흥로 490-1 형식
            road = address_letter[2] #도로명 설정

            if '-' in address_letter[3]: #건물번호 부번이 있을 경우
                building_no = address_letter[3].split('-')
                building_no = building_no[0]
            else:
                building_no = address_letter[3]
        else: #서울특별시 종로구 세종로 세종대로 147-2 형식
            road = address_letter[3] #도로명 설정

            if '-' in address_letter[4]: #건물번호 부번이 있을 경우
                building_no = address_letter[4].split('-')
                building_no = building_no[0]
            else:
                building_no = address_letter[4]    
        

        ###### 인접 도로주소명 찾아보기 ######
        j = 1
        while j < int(building_no):
            new_build_no_plus = int(building_no) + j
            new_build_no_minus = int(building_no) - j

            place = post_df[(post_df['시군구'] == gu) & (post_df['도로명'] == road) & (post_df['건물번호본번'] == int(new_build_no_plus))]
            if len(place) == 0: #장소가 없어- 일단 결측치 채워넣어. 지하철 데이터는 인촌로인데 우편번호는 고려대로야...
                pass
            else:
                place = place.iloc[0]
                if pd.isna(place['행정동명']) : #### 행정동명이 없을 경우에는 같은 우편번호의 행정동을 가져오자, 
                                                   #행정동이 여러개일 때는 최빈값 가져오기
                    postno = place['우편번호']
                    postno_df = post_df[post_df['우편번호'] == postno]
                    postno_df = list(postno_df['행정동명'])
                    newlist = [x for x in postno_df if pd.isnull(x) == False and x != 'nan']
                    cnt = Counter(newlist)
                    mode = cnt.most_common(1)
                    if mode == []:
                        bus_df.loc[i, '행정동'] = np.nan
                    else:
                        name = mode[0][0]  
                        bus_df.loc[i, '행정동'] = name

                else:  #### 행정동명이 있을 경우에는 그냥 행정동명을 삽입
                    name = place['행정동명']
                    bus_df.loc[i, '행정동'] = name
                break


            place = post_df[(post_df['시군구'] == gu) & (post_df['도로명'] == road) & (post_df['건물번호본번'] == int(new_build_no_minus))]
            if len(place) == 0: #장소가 없어- 일단 결측치 채워넣어. 지하철 데이터는 인촌로인데 우편번호는 고려대로야...
                pass

            else:
                place = place.iloc[0]
                if pd.isna(place['행정동명']) : #### 행정동명이 없을 경우에는 같은 우편번호의 행정동을 가져오자, 
                                                   #행정동이 여러개일 때는 최빈값 가져오기
                    postno = place['우편번호']
                    postno_df = post_df[post_df['우편번호'] == postno]
                    postno_df = list(postno_df['행정동명'])
                    newlist = [x for x in postno_df if pd.isnull(x) == False and x != 'nan']
                    cnt = Counter(newlist)
                    mode = cnt.most_common(1)
                    if mode == []:
                        bus_df.loc[i, '행정동'] = np.nan
                    else:
                        name = mode[0][0]  
                        bus_df.loc[i, '행정동'] = name

                else:  #### 행정동명이 있을 경우에는 그냥 행정동명을 삽입
                    name = place['행정동명']
                    bus_df.loc[i, '행정동'] = name 
                break
            j = j+1
```
