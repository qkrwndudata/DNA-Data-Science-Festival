```python
import pandas as pd
```

```python
final = pd.read_csv("/content/drive/MyDrive/DNA/Data Archive/QGIS/버퍼 별 데이터/FINAL_SCORE.csv", encoding = 'euc-kr')
final['범죄율'] = -final['범죄율']
final['경사도 비율'] = 1 - final['경사도 비율']
final
```

```python
label_df = final[['주소', '구', '동']]
final_df1 = final[['버스정류장', '소방서', '지하철', '전통시장', '병의원', '공원면적', '독거노인비율', '범죄율', '사회참여율', '경사도 비율']]
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler()
final_df2 = min_max_scaler.fit_transform(final_df1)
final_df2 = pd.DataFrame(final_df2, columns = list(final_df1.columns))
```

```python
final_df2 = pd.concat([label_df, final_df2], axis = 1)
final_df2
```

```python
final_df2['Scoring'] = 0.07702755*final_df2['전통시장'] + 0.02971888*final_df2['소방서'] + 0.07736474*final_df2['범죄율'] + 0.01501857*final_df2['공원면적'] + 0.02716272*final_df2['사회참여율'] + 0.15675841*final_df2['병의원'] + 0.24617620*final_df2['독거노인비율'] + 0.02439428*final_df2['지하철'] + 0.02439428*final_df2['버스정류장']  + 0.03689874*final_df2['경사도 비율'] 
final_df2['Scoring'] = final_df2['Scoring']/0.7149144
final_df2.sort_values(by='Scoring', ascending=False, inplace=True)
final_df2.reset_index(drop = True, inplace = True)
```
